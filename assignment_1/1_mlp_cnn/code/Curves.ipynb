```python
import numpy as np
import torch
import pickle
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.rc('axes', labelsize=15)
plt.rc('font', size=15)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
```

# MLP Numpy

```python
performance = pickle.load(open("MLP_numpy_curves.p", "rb"))

# def smooth(x, N):
#     cumsum = np.cumsum(np.insert(x, 0, 0)) 
#     return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
train_loss = np.array(performance['Training loss (batch)'])
test_loss = np.array(performance['Test loss'])

plt.plot(train_loss[:, 0], train_loss[:, 1], label='Training loss (batch)', color='dodgerblue', zorder=1)

plt.plot(test_loss[:, 0], test_loss[:, 1], color='indigo', zorder=2)
plt.scatter(test_loss[:, 0], test_loss[:, 1], label='Test loss', color='indigo', marker='X', s=70, zorder=2)
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
train_acc = np.array(performance['Training accuracy (batch)'])
test_acc = np.array(performance['Test accuracy'])

plt.plot(train_acc[:, 0], train_acc[:, 1], color='dodgerblue')
plt.scatter(train_acc[:, 0], train_acc[:, 1], color='dodgerblue', marker='X', s=70, label='Training accuracy (batch)')

plt.scatter(test_acc[:, 0], test_acc[:, 1], label='Test accuracy', color='indigo', marker='X', s=70)
plt.plot(test_acc[:, 0], test_acc[:, 1], color='indigo')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
