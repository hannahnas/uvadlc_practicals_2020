```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update(plt.rcParamsDefault)
plt.rc('axes', labelsize=15)
plt.rc('font', size=15)
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
def plot_performance(data):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    ax = ax.ravel()
    
    acc = np.array(data['Accuracy'])
    loss = np.array(data['Loss'])
    
    ax[0].plot(loss[:, 0], loss[:, 1])
    ax[1].plot(acc[:, 0], acc[:, 1])

    ax[0].legend()
    ax[0].set_xlabel('Step')
    ax[0].set_ylabel('Train loss')
    ax[1].set_xlabel('Step')
    ax[1].set_ylabel('Train accuracy')
#     ax[1].legend()
    plt.show()
    
```

# LSTM

```python
grimm = pickle.load(open("./Performance/SpeechGeneration_data_book_EN_grimms_fairy_tails.p", "rb"))

plot_performance(grimm)
```
