```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update(plt.rcParamsDefault)
plt.rc('axes', labelsize=15)
plt.rc('font', size=15)
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
def mean_std_performance(data):
    losses = []
    accs = []

    for performance in data:
        loss, acc = np.array(performance['Loss'])[:400,1], np.array(performance['Accuracy'])[:400,1]
        losses.append(loss)
        accs.append(acc)
        
    mean_loss = np.mean(losses, axis=0)
    std_loss = np.std(losses, axis=0)
    mean_acc = np.mean(accs, axis=0)
    std_acc = np.std(accs, axis=0)
    
    return mean_loss, std_loss, mean_acc, std_acc

def plot_performance(data):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    ax = ax.ravel()
    
    time = ["$\mu$ for t=5", "$\mu$ for t=10", "$\mu$ for t=15"]
    colors = ["mediumblue", "#009E73", "#E69F00"]
    iterate = zip(data, time, colors)
    for performance, seq_len, color in iterate:
        mean_loss, std_loss, mean_acc, std_acc = mean_std_performance(performance)
        x = np.arange(len(mean_loss))
        # Plot loss
        ax[0].plot(x, mean_loss, label=seq_len, color=color)
        ax[0].fill_between(x, mean_loss-std_loss, mean_loss+std_loss, alpha=0.6, label='$\sigma$', color=color)
        # Plot accuracy             
        ax[1].plot(np.arange(400), mean_acc, label=seq_len, color=color)
        ax[1].fill_between(x, mean_acc-std_acc, mean_acc+std_acc, alpha=0.6, label='$\sigma$', color=color)
    ax[0].legend()
    ax[0].set_xlabel('Step')
    ax[0].set_ylabel('Train loss')
    ax[1].set_xlabel('Step')
    ax[1].set_ylabel('Train accuracy')
    ax[1].legend()
    plt.show()
    
```
